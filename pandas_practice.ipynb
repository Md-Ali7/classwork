###Using groupby(), plot the total number of medals awarded at each of the Olympic games throughout history. (2 points)

```python
df_year_medal = df.groupby(['Edition'])['Medal'].count().reset_index()
df_year_medal
```

```python
#creating shades of color for bar graph
cmap = get_cmap('Greens')
bar_color = cmap(df_year_medal['Medal'] / df_year_medal['Medal'].max())

#ploting graph
plt.barh(df_year_medal['Edition'],df_year_medal['Medal'],color=bar_color,height = 2)
plt.ylabel('Edition')
plt.xlabel('Number of Medals')
plt.title('Number of Medals by Edition')

plt.yticks(df_year_medal['Edition'])



plt.show()
```

###Create a list showing the total number of medals won for each country over the history of the Olympics. For each country, include the year of the first and most recent Olympic medal wins. (3 points)

```python
#getting list of all countries with their medals
df_country = df.groupby('NOC').agg(
    Total_Medals=pd.NamedAgg(column='Medal', aggfunc='count'),
    First_Medal_Year=pd.NamedAgg(column='Edition', aggfunc='min'),
    Most_Recent_Medal_Year=pd.NamedAgg(column='Edition', aggfunc='max')
).reset_index().sort_values(by=['Total_Medals'],ascending=False)
df_country
```

###Plot the number of medals achieved by the Chinese team (men and women) in Beijing 2008 using:
- matplotlib
- Seaborn      (3 points)

```python
CHN_medals = df[(df['NOC']=='CHN')&(df['Edition']==2008)].groupby(['Gender'])['Medal'].count().reset_index()
CHN_medals
```

```python
bar_colors = ['skyblue', 'salmon']
#ploting the graph
plt.bar(CHN_medals['Gender'],CHN_medals['Medal'],color=bar_colors)
plt.xlabel('Gender')
plt.ylabel('Number of Medals')
plt.title('Number of Medals by Gender(CHN-2008)')
plt.show()
```

```python
sns.barplot(data= CHN_medals,x="Gender",y="Medal").set(title="Number of Medals by Gender(CHN-2008)")
```
